```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
tree_reg = DecisionTreeRegressor(max_depth=2)
```

```python
X = np.linspace(-1, 1, 100).reshape(100, 1)
Y = 2*X**2  + X + 0.5
plt.scatter(X, Y)
```

```python
noise = np.random.normal(0, 0.1, X.shape)
Y = Y + noise
plt.scatter(X, Y)
```

```python
tree_reg.fit(X, Y)
```
